## 3. Strategic Recommendations

1. **Congestion Management**: With a Cohen's d of -2.75, Congestion is confirmed as a massive driver. Prioritize backhaul expansion in High-Utilization Zones.
2. **Latency Reduction**: Feature importance confirms Latency is the top predictor. Optimize routing table and edge cache placement.
3. **Proactive Maintenance**: Use Anomaly detection clusters to trigger field inspections before major outages (e.g., 5 PM peaks).
